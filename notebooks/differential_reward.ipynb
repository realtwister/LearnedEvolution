```python
def plt_reward(reward_fn, fitness):
    rewards =[];
    cum_rewards = [];
    cum_reward = 0;
    reward_fn.reset();
    for f in fitness:
        reward = reward_fn([],f);
        rewards.append(reward);
        cum_reward += reward;
        cum_rewards.append(cum_reward);
    plt.figure();
    plt.subplot(3, 1, 1)
    plt.plot(fitness, 'o', label="fitness");
    plt.ylabel('fitness')
    plt.subplot(3, 1, 2)
    plt.plot(rewards, 'o', label="reward");
    plt.ylabel('reward');
    plt.subplot(3, 1, 3)
    plt.plot(cum_rewards, 'o', label="cumulative reward");
    plt.ylabel('cumulative reward');
    plt.show();
```

```python
import numpy as np;

class LinearReward:
    def reset(self):
        self.mean_fs =[];
    def __call__(self, population, fitness):
        mean_fitness = np.mean(fitness);
        self.mean_fs.append(mean_fitness);
        return mean_fitness-1;

class DifferenceReward:
    def reset(self):
        self.mean_fs = [];
    
    def __call__(self, population, fitness):
        mean_fitness = np.mean(fitness);
        self.mean_fs.append(mean_fitness);
        if len(self.mean_fs)<2:
            return 0;
        return self.mean_fs[-1] - self.mean_fs[-2]-1;

class RatioReward:
    def reset(self):
        self.mean_fs = [];
    
    def __call__(self, population, fitness):
        mean_fitness = np.mean(fitness);
        self.mean_fs.append(mean_fitness);
        if len(self.mean_fs)<3:
            return 0;
        numerator = self.mean_fs[-1] - self.mean_fs[-2];
        denominator = max(np.abs(self.mean_fs[-2]-self.mean_fs[-3]), 1e-10);
        return numerator/denominator;
        
        
```

```python
import numpy as np;

div = DivergencePenalty();
dif = DifferentialReward();

class Reward:
    def reset(self):
        div.reset();
        dif.reset();
    def __call__(self, population, fitness):
        return 0.5 * div(population,fitness)+ dif(population, fitness);

fitness= np.linspace(20,120, 100);
plt_reward(LinearReward(), fitness);
plt_reward(DifferenceReward(), fitness);
plt_reward(RatioReward(), fitness);
```

```python
fitness = [0];
gamma = 0.9;
max_value = 1/(1-gamma);
for i in range(50):
    fitness.append(fitness[-1]+gamma**i);
    
for i, f in enumerate(fitness):
    fitness[i] = fitness[i]/max_value;
plt_reward(LinearReward(), fitness);
plt_reward(DifferenceReward(), fitness);
plt_reward(RatioReward(), fitness);

print(1/(1-gamma))
    
```
